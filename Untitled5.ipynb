```python
from os import listdir
from os.path import isfile, join
import numpy as np
import cv2
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
%matplotlib inline
#C:/Users/Shubham/Documents/Image_Processing/Dataset/Train/color/
```

```python
mypath='C:/Users/Shubham/Documents/Image_Processing/Dataset/Train/color/'
onlyfiles = [ f for f in listdir(mypath) if isfile(join(mypath,f)) ]
images = np.empty(len(onlyfiles), dtype=object)
for n in range(0,len(onlyfiles ):
    images[n] = cv2.imread( join(mypath,onlyfiles[n]))
    r,c=images[n].shape[:2]
    out_r= 120
    images[n] = cv2.resize(images[n],(int(out_r*float(c)/r), out_r))
    images[n] = cv2.cvtColor(images[n], cv2.COLOR_BGR2RGB)
```

```python
km=KMeans(n_clusters=5)
for n in range(0,len(onlyfiles)):
    km.fit(images[n].reshape(-1,3))
    colors=np.asarray(km.cluster_centers_,dtype='uint8')
    per=np.asarray(np.unique(km.labels_,return_counts=True)[1],dtype='float32')
    per=per/(images[n].reshape(-1,3)).shape[0]
    dom=[[per[ix],colors[ix]] for ix in range(colors.shape[0])]
    Dom=sorted(dom, key =lambda z:z[0],reverse=True)
    start=0
    j,k = 0,0
    for i in range(km.n_clusters):
        print(Dom[i][1]) 
        if int(Dom[i][1][0]/2) == int(Dom[i][1][1]/2) == int(Dom[i][1][2]/2):
            j += 1
        else:
            k += 1
    if j > k:
        print('grey')
        plt.imshow(images[n])
        plt.pause(0.0001)
    if j < k:
        print ('color')
        plt.imshow(images[n])
        plt.pause(0.0001)
```
